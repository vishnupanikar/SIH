Importing the shapefile for India

```python
state_coordinates = shp.Reader('Igismap/Indian_States.shp')
```

```python
state_coordinates.fields
```

```python
state_coordinates.records()
```

```python
state_coordinates.shapes()[18].points
```

```python
def create_dataframe(sf):
    column = [i[0] for i in sf.fields[1:]]
    State =  [i[0] for i in sf.records()]
    loc = [i.points for i in sf.shapes()]
    
    state_sf = pd.DataFrame(data = State , columns=column)
    state_sf = state_sf.assign(coordinates = loc)
    return state_sf
```

```python
state_sf = create_dataframe(state_coordinates)
state_sf.head(5)
```

```python
state_sf.rename({'st_nm':'State'},axis = 1 , inplace=True)
```

```python
state_sf.head(3)
```

```python
state_sf.State[1] = 'Arunachal Pradesh'
state_sf.State[12] = 'Jammu and Kashmir'
state_sf.State[23] = 'Delhi'
```

```python
state_df = pd.merge(state_data,state_sf,how='inner',on='State')
```

```python
state_df.head(5)
```

```python
state_df = state_df[['State','Confirmed','Recovered','Deaths','Active','State_code','coordinates']]
```

```python
palette = tuple(reversed(palette))
```

```python
longs = []
lat = []
for s in state_df.coordinates:
    long = np.mean([loc[0] for loc in s])
    lats = np.mean([loc[1] for loc in s])
    longs.append(long)
    lat.append(lats)
```

```python
State_name = [i for i in state_df.State]
values = [i for i in state_df.Active]
```

```python
color_mapper = LogColorMapper(palette = palette)
```

```python
data = dict(x = longs , y = lat , name = State_name , active = values)
```

```python
TOOLS = 'pan,wheel_zoom,reset,hover,save'
```

```python
p = figure(
    title="Active COVID19 cases", tools=TOOLS,
    x_axis_location=None, y_axis_location=None,
    tooltips=[
        ("Name", "@name"), ("Active Cases", "@active")
    ])
```

```python
p.grid.grid_line_color = None
p.hover.point_policy = "follow_mouse"
```

```python
p.patches('x', 'y', source=data,
          fill_color={'field': 'active', 'transform': color_mapper},
          fill_alpha=0.7, line_color="white", line_width=0.5)
```

```python
show(p)
```
